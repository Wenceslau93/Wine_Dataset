# Wine Dataset

<br/>

Esses dados são resultados de uma análise química de vinhos cultivados em uma região da Itália, derivados de três cultivares diferentes.

Objetivo do estudo, é realizar previsão se o vinho é tinto ou branco.

```python
# Importanto as bibliotecas

import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.ensemble import ExtraTreesClassifier

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
```

```python
# Carregando o dataset

arquivo = pd.read_csv('wine_dataset.csv')
```

```python
# Visualizar o tamanho do dataset

arquivo.shape
```

```python
# Visualizar as primeiras linhas

arquivo.head()
```

```python
# Visualizar as últimas linhas 
    
arquivo.tail()
```

```python
# Trocando os valores da variável target style, para númerico

arquivo['style'] = arquivo['style'].replace('red', 0)
arquivo['style'] = arquivo['style'].replace('white', 1)
```

```python
# Visualizar as primeiras linhas após a substituição dos valores da variável númerica

arquivo.head()
```

```python
# Visualizar as últimas linhas após a substituição dos valores da variável númerica

arquivo.tail()
```

```python
# Separando as variáveis preditoras que ficaram armazenadas na nova variável x

x = arquivo.drop('style', axis = 1)
```

```python
# Separando as variável alvo que ficara armazenada em y

y = arquivo['style']
```

```python
# conjunto de dados para treino e teste. Será utilizado 30% dos dados para teste e 70% para treino

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3)
```

```python
# Modelo utilizado para o algoritmo para criar as árvores de decisão

modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino)
```

```python
# Imprimindo o resultado

resultado = modelo.score(x_teste, y_teste)
print("Acurácia de:", resultado * 100)
```
